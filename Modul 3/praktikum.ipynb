```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA

# 1. Data Cleaning

# Memuat dataset
df = pd.read_csv('smartphones.csv')
display(df)

# Penggantian nilai NaN dengan nilai 0
df.fillna(0, inplace=True)

print("\nJumlah nilai (NaN) setelah penggantian:")
display(df)

# # 2. Encoding 
label_encoder = LabelEncoder()
df['model'] = label_encoder.fit_transform(df['model'])

print()

# # Menampilkan informasi mengenai data setelah data cleaning dan encoding
print("Informasi dataset setelah Data Cleaning dan Encoding:")
display(df.head(5))

# # Memisahkan kolom target
target = df['model']
features = df.drop(columns=['5G_or_not', 'model'])

# print()

# Memisahkan features dan target
X = df.drop(columns=['5G_or_not', 'model'])  # Menghapus kolom UserID (jika tidak digunakan sebagai fitur) dan kolom target 'Beli'
y = df['model']  # Menyimpan kolom target 'Beli'

# Data Splitting: Membagi data menjadi set pelatihan dan pengujian dengan proporsi 70:30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Data fitur set pelatihan:")
display(X_train.head(5))

print("\nData fitur set pengujian:")
display(X_test.head(5))

print("\nTarget set pelatihan:")
display(y_train.head(5))

print("\nTarget set pengujian:")
display(y_test.head(5))


# Normalisasi data menggunakan StandardScaler
scaler = StandardScaler()
df[['model', 'price']] = scaler.fit_transform(df[['model', 'price']])
display(df[['model', 'price']])

min_max_scaler = MinMaxScaler()
df_min_max_scaled = min_max_scaler.fit_transform(df[['price', 'processor_speed', 'ram_capacity']])
df_min_max_scaled = pd.DataFrame(df_min_max_scaled, columns=['price', 'processor_speed', 'ram_capacity'])

print(df_min_max_scaled.head(5))

df_min_max_scaled.to_csv('smartphones_scaled.csv', index=False)

# data split
X = df.drop(columns=['price'])
y = df['processor_speed']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Jumlah data dalam set pelatihan:", X_train.shape[0])
print("Jumlah data dalam set pengujian:", X_test.shape[0])

# pca
X = df.drop(columns=['brand_name'])

X = pd.get_dummies(X)

X.fillna(X.mean(), inplace=True)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)

print("Jumlah Komponen PCA yang Dipilih:", pca.n_components_)


print("Ukuran Data Setelah Reduksi Dimensi:", X_pca.shape)

plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.5)
plt.title('Scatter Plot Hasil PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# result_train.to_csv('training_preprocessed.csv', index=False)
# result_test.to_csv('testing_preprocessed.csv', index=False)
```
