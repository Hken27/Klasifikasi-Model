```python
# Import library
import pandas as pd
from scipy.stats import probplot
import matplotlib.pyplot as plt


# Membaca dataset
data = pd.read_csv('smartphones.csv')

print ("==business problem==")
print (" Apa tujuan dari analisis imi? ")
print (" Apa masalah utama yang ingin dipecahkan? ")
print (" Siapa yang menggunakan hasil analisis ini?  ")
print (" Apa sumber data yang tersedia dan seberapa reliable? ")
print (" Bagaimana data dapat diolah untuk memberikan wawasan yang berguna?  ")
print()
print()

# Analisis deskriptif statistik
deskripsi_statistik = data.describe()
print(deskripsi_statistik)


# Visualisasi box plot untuk kolom numerik
kolom_numerik = data.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12, 6))
data[kolom_numerik].boxplot()
plt.title('Box plot untuk Kolom Numerik')
plt.xticks(rotation=45)
plt.show()

# Membuat box plot untuk harga
plt.boxplot(data['price'])
plt.title('Box Plot Harga')
plt.show()


# Membuat Q-Q plot untuk harga
probplot(data['price'], plot=plt)
plt.title('Q-Q Plot Harga')
plt.show()

```
