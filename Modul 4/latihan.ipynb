```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report

# Muat dataset
data = pd.read_csv('iklan_sosmed.csv')
display(data.head(5))

X = data.drop('Beli', axis=1)
y = data['Beli']

# Encoding kolom 'Kelamin'
# Dan normalisasi
preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(), ['Kelamin']),
        ('std_scaler', StandardScaler(), ['Usia', 'EstimasiGaji'])
    ],
    remainder='passthrough')

X_transformed = preprocessor.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X_train, y_train)

# Lakukan prediksi menggunakan model RandomForestClassifier
y_pred_rf = rf_model.predict(X_test)

print("Random Forest Klasifikasi:")
print(classification_report(y_test, y_pred_rf, zero_division=1)) 

svm_model = SVC(kernel='linear', random_state=42)

svm_model.fit(X_train, y_train)

y_pred_svm = svm_model.predict(X_test)

# Evaluasi model SVM Classifier
print("\nSupport Vector Machine Klasifikasi:")
print(classification_report(y_test, y_pred_svm, zero_division=1))
```
